```python
intents = {
    "greeting": {
        "patterns": ["hi", "hello", "hey"],
        "responses": ["Hello!", "Hi there!", "Hey!"]
    },
    "price_query": {
        "patterns": ["price", "cost", "rate"],
        "responses": ["Please tell me the commodity name."]
    },
    "commodity_price": {
        "patterns": ["tomato", "potato", "onion", "carrot", "chilli"],
        "responses": ["Here is the price:"]
    },
    "goodbye": {
        "patterns": ["bye", "exit"],
        "responses": ["Goodbye!", "See you later!"]
    }
}
commodity_prices = {
    "tomato": 20,
    "potato": 10,
    "onion": 30,
    "carrot": 20,
    "chilli": 10
}
```

```python
import random
def get_intent(user_input):
    user_input=user_input.lower()
    for intent, data in intents.items():
        for pattern in data["patterns"]:
            if pattern in user_input:
                return intent
    return "unknown"


def get_commodity(user_input):
    user_input=user_input.lower()
    for item in commodity_prices.keys():
        if item in user_input:
            return item
    return None


def chatbot():
    while True:
        user_input=input("User: ")
        commodity=get_commodity(user_input)

        if commodity:
            price=commodity_prices[commodity]
            print(f"Bot: The price of {commodity} is ₹{price} per kg.")
            continue

        intent=get_intent(user_input)

        if intent == "unknown":
            print("Bot: I didn't understand.")

        elif intent == "price_query":
            item = input("Bot: Which commodity? ").lower()
            if item in commodity_prices:
                print(f"Bot: The price of {item} is ₹{commodity_prices[item]} per kg.")
            else:
                print("Bot: Sorry, I don't have that commodity.")

        else:
            response=random.choice(intents[intent]["responses"])
            print("Bot:", response)

        if intent == "goodbye":
            break


chatbot()
```
